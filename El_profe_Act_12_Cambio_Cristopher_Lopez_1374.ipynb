## las variables
1. CP -->  Costo del producto
2. CD -->  Cantidad de dinero
3. C  -->  Cambio

## Formulas
$Cambio = CD - CP$


## Diagrama de flujo Lucidchart
<img src="https://raw.githubusercontent.com/CristopherLopez67/Calcular-cambio-if-else-/refs/heads/main/DF%20if%20else%20cambio2.png" alt="Descripción" width="350" height="500">

```python

# para cuando es verdadero (si)
# calculando el cambio usando if else
# Datos de entrada CD, CP
# mostrar cambio si es falso
# si es verdadero mostrar no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD < CP:
    print("No alcanza")
else:
    C = CD - CP
    print("El cambio es: ", C)
    print("Fin del programa")
print("origrama realizado por cristopher lopez NL = 30")
```

```python
# para cuando es falso (no)
# calculando el cambio usando if else
# Datos de entrada CD, CP
# mostrar cambio si es falso
# si es verdadero mostrar no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD < CP:
    print("No alcanza")
else:
    C = CD - CP
    print("El cambio es: ", C)
    print("Fin del programa")
print("origrama realizado por cristopher lopez NL = 30")
```
